```python
# Tendo um conjunto de dados grande eu quero filtrar, mas usando sql
```

```python
from pyspark.sql import SparkSession
```

```python
contexto = SparkSession.builder.appName("Acoes").getOrCreate()
```

```python
contexto
```

```python
# inferSchema permite inferir os tipos de dados e o poder das consultas aumenta 
acoes = contexto.read.csv("appl_stock.csv", header=True, inferSchema=True)
```

```python
acoes.show()
```

```python
acoes.printSchema()
```

```python
# Quero retornar o conjunto de dados onde o Valor de Fechamento é menor que 500
acoes.filter("Close<500").show()
```

```python
# Quero retornar o conjunto de dados onde o Valor de Fechamento é menor que 500 e mostre apenas a coluna Open

acoes.filter("Close<500").select("Open").show()
```

```python
# Quero retornar o conjunto de dados onde o Valor de Fechamento(Close) é menor que 700 e mostre apenas as colunas Open e Date

acoes.filter("Close<700").select(['Date', 'Open']).show()
```

```python
acoes.filter( acoes["Close"]<800  ).show()
```

```python
# Agora eu gostaria do valor da coluna Close <300 E  Open>500
resultado = acoes.filter( (acoes["Close"]<300) & (acoes["Open"]>500) )
resultado.show()
```

```python
resultado = acoes.filter( (acoes["Close"]<300) & (acoes["Open"]>50) )
resultado.show()
```

```python
# Quero a linha 
resultado = acoes.filter( acoes["Low"] == 198.699995 ).collect()
```
