```python
from pyspark.sql import SparkSession
```

```python
contexto = SparkSession.builder.appName("datas").getOrCreate()
```

```python
datas = contexto.read.csv("appl_stock.csv", header=True, inferSchema=True)
```

```python
datas.show()
```

```python
datas.printSchema()
```

```python
from pyspark.sql.functions import dayofmonth
```

```python
datas.select(dayofmonth(datas["Date"])).show()
```

```python
from pyspark.sql.functions import month, year, dayofyear
```

```python
datas.select(month(datas["Date"])).show()
```

```python
datas.select(year(datas["Date"])).show()
```

```python
datas.select(dayofyear(datas["Date"])).show()
```

```python
novodf = datas.withColumn("Ano", year(datas["Date"]))
```

```python
novodf.groupBy("Ano").mean().show()
```

```python
novodf.groupBy("Ano").mean()[['Ano','avg(High)', 'avg(Low)']].show()
```
